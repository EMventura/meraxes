The purpose of this notebook is to compare the results of Meraxes, using the GiggleZ_MR horizontal trees, and the Mutch+ 2013 Formation History model using the equivalent vertical trees.  The FHmodel (vertical trees) provides a decent z=0 SMF, even without calibrating.  The Meraxes result (with the horizontal trees) is terrible though, and over produces low mass galaxies.  Hopefully this comparison will allow us to find out what is going wrong with Meraxes.

```python
# constants
MERAXES_FILE = "/home/smutch/data/meraxes/GiggleZ_MR_no_ghost_test/output/GiggleZ_MR_no_ghost_test-fullhist.hdf5" 
FHMODEL_FILE = "/home/smutch/data/gigglez/vertical/test-20130726/output/GiggleZ.hdf5"
```

The most bound particle IDs indicate that these are indeed the same halos.  However, they have different properties, including the number of particles.  I don't see how this could be?  The Mvir values are actually the same though, with the differences just being due to differing Hubble constants.  The velocities are also considerably different.

The stellar masses of the galaxies in these halos shows massive variation with the FHmodel stellar mass being >x2 that of the Meraxes galaxy.

What about the HMFs at z=0?

## Halo Mass Functions

```python
# Construct the halo mass function for each model at z=0 and compare
def mass_func(masses, volume, bins=30):
    """Construct the halo mass function."""
    counts, edges = histogram(masses, bins=bins, range=(10,15))
    bin_width = edges[1]-edges[0]
    centers = edges[:-1] + (bin_width*0.5)
    phi = counts/bin_width/volume
    return centers, phi

m_type0_sel = m_gals["Type"]==0
m_hmf = mass_func(np.log10(m_gals[m_type0_sel]["Mvir"]*1.e10), m_simprops["Volume"])
fh_type0_sel = fh_gals["Type"]==0
# Note the hubble constant correction here...
fh_hmf = mass_func(np.log10(fh_gals[fh_type0_sel]["Mvir"]*1.e10), fh_simprops["Volume"])
diff_hmf = (m_hmf[0], (m_hmf[1]-fh_hmf[1])/fh_hmf[0])

fig, axes = subplots(1,2, figsize=(8*2, 6))
hmf_xlabel = r"$\log_{10}(M_{\rm vir}/\rm{M_{\odot}})$"
ax = axes[0]
ax.set_yscale('log')
ax.plot(m_hmf[0], m_hmf[1], label="Meraxes")
ax.plot(fh_hmf[0], fh_hmf[1], label="FHmodel")
ax.text(0.95, 0.95, "z=0, type=0", horizontalalignment="right", verticalalignment="top", transform=ax.transAxes)
ax.legend(loc="lower left")
ax.set_ylabel(r"$\phi [{\rm dex^{-1}Mpc^{-3}}]$")
ax.set_xlabel(hmf_xlabel)
ax = axes[1]
ax.plot(diff_hmf[0], diff_hmf[1], label="fractional difference")
#ax.set_yscale('log')
ax.set_ylabel("frac")
ax.set_xlabel(hmf_xlabel)
ax.text(0.95, 0.95, "z=0, type=0", horizontalalignment="right", verticalalignment="top", transform=ax.transAxes)
ax.legend(loc="lower left")
tight_layout()
draw()
```

So the FOF HMF looks fine, with only very minor variations between models.

## Histories

Let's check the Mvir histories of the most massive halos in both models.

Note that I have adapted both models especially to provide useful indices for tracking galaxies across timesteps.
These are *not* the most efficient tracking indices by any means, however, they work transparently which is key for finding bugs!

```python
# We know that we have all of the snapshots in the output, so there is no need to read them
snaps = arange(59)

# Now lets construct the galaxy histories

# Meraxes
target_ID = m_gals["ID"][meraxes_mmind]
m_galhist = []
m_snaps = []
m_nsatshist = []
m_ntype2 = []
for snap in snaps:
    gals = read_meraxes_gals(MERAXES_FILE, snapshot=snap, verbose=False)
    sel = gals["ID"]==target_ID
    if sel.any():
        m_galhist.append(gals[sel])
        sel = gals["CentralGal"]==gals[sel]["CentralGal"]
        m_nsatshist.append(count_nonzero(sel)-1)
        sel = gals[sel]["Type"]==2
        m_ntype2.append(count_nonzero(sel))
        m_snaps.append(snap)
m_galhist = np.array(m_galhist, dtype=m_gals.dtype)
m_nsatshist = np.array(m_nsatshist, int)
m_ntype2 = np.array(m_ntype2, int)
m_snaps = np.array(m_snaps, int)

# FHmodel
target_ID = fh_gals["ID"][fh_mmind]
target_tree = fh_gals["TreeIndex"][fh_mmind]
fh_galhist = []
fh_ntype2 = []
fh_snaps = []
fh_nsatshist = []
for snap in snaps:
    gals = read_fhmodel_gals(FHMODEL_FILE, snapshot=snap, verbose=False)
    sel = (gals["ID"]==target_ID) & (gals["TreeIndex"]==target_tree)
    if sel.any():
        fh_galhist.append(gals[sel])
        sel = (gals["FOFHaloIndex"]==gals[sel]["FOFHaloIndex"]) & (gals["TreeIndex"]==gals[sel]["TreeIndex"])
        fh_nsatshist.append(count_nonzero(sel)-1)
        sel = gals[sel]["Type"]==2
        fh_ntype2.append(count_nonzero(sel))
        fh_snaps.append(snap)
fh_galhist = np.array(fh_galhist, dtype=fh_gals.dtype)
fh_nsatshist = np.array(fh_nsatshist)
fh_ntype2 = np.array(fh_ntype2, int)
fh_snaps = np.array(fh_snaps, int)
```

```python
# Now lets plot them!
fig, axes = subplots(7,1, figsize=(8*2, 24))
i_ax = 0

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, log10(m_galhist["Mvir"]*1.e10), label="Meraxes")
ax.plot(fh_snaps, log10(fh_galhist["Mvir"]*1.e10), label="FHmodel", alpha=0.5)
ax.set_ylabel(hmf_xlabel)
ax.set_xlabel("snapshot")
ax.legend(loc="upper left")

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, log10(m_galhist["dMdt"]), label="Meraxes")
ax.plot(fh_snaps, log10(fh_galhist["dMdt"]), label="FHmodel", alpha=0.5)
ax.set_ylabel(r"$\log_{10}(\frac{dM_{\rm vir}}{dt}/[\rm{M_{\odot}/yr}])$")
ax.set_xlabel("snapshot")
ax.legend(loc="upper left")

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, (m_galhist["StellarMass"]*1.e10), label="Meraxes")
ax.plot(fh_snaps, (fh_galhist["StellarMass"]*1.e10), label="FHmodel", alpha=0.5)
ax.set_ylabel(r"$\log_{10}(M_{*}/\rm{M_{\odot}})$")
ax.set_xlabel("snapshot")
ax.legend(loc="upper left")

ax = axes[i_ax]
i_ax+=1
lm, = ax.plot(m_snaps, m_nsatshist, label="Meraxes")
lfh, = ax.plot(fh_snaps, fh_nsatshist, label="FHmodel", alpha=0.5)
ax.plot(m_snaps, m_nsatshist-m_ntype2, ls=":", color=lm.get_color(), label="Type 1")
ax.plot(fh_snaps, fh_nsatshist-fh_ntype2, ls=":", color=lfh.get_color(), alpha=0.5)
ax.plot(m_snaps, m_ntype2, ls="--", color=lm.get_color(), label="Type 2")
ax.plot(fh_snaps, fh_ntype2, ls="--", color=lfh.get_color(), alpha=0.5)
ax.set_ylabel("N satellites")
ax.set_xlabel("snapshot")
ax.legend(loc="upper left")

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, m_galhist["Type"], label="Meraxes")
ax.plot(fh_snaps, fh_galhist["Type"], label="FHmodel", alpha=0.5)
ax.set_ylabel("Type")
ax.set_xlabel("snapshot")
ax.legend(loc="center left")
ax.axis((0,60,-0.1,2.1))

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, m_galhist["id_MBP"], label="Meraxes")
ax.plot(fh_snaps, fh_galhist["MostBoundID"], label="FHmodel", alpha=0.5)
ax.set_ylabel("ID of most bound particle")
ax.set_xlabel("snapshot")
ax.legend(loc="center left")

ax = axes[i_ax]
i_ax+=1
ax.plot(m_snaps, m_galhist["Vmax"], label="Meraxes")
ax.plot(fh_snaps, fh_galhist["Vmax"], label="FHmodel", alpha=0.5)
ax.set_ylabel(r"$V_{\rm max}[{\rm km\,s^{-1}}]$")
ax.set_xlabel("snapshot")
ax.legend(loc="center left")

tight_layout()
draw()
```

This is weird.  There are definite differences in the halo histories between the vertical and horizontal trees.  This might be because Greg is using the most massive halo or De Lucia criterion for picking the first progenitor halo in the vertical trees.  In the horizontal trees, the match score is used I think.

What really doesn't make sense is that the Mvir values and dMvir/dt values are identical for the haloes of the galaxies in both models, yet there is a large variation in the stellar mass growth histories.  If we assume that both sets of trees are right (seems likely if the halo mass growth histories agree) then this difference in stellar mass must be coming from mergers.  In the FHmodel, there appears to be far more stellar mass growth due to mergers.  The number of satellites in each FOFHalo shows an increasing disparity as time goes on.  This probably supports the assertion that mergers are the problem here.

Further investigation shows that the number of Type 1 satellites is identical between models as one would hope.  It is the number of type 2 sats that are causing the problem.  This - I think - pins down definitively that there is something funky going on with the merger clocks in Meraxes...
